    In E-commerce, both the retailers (here brands) and the company have to make profit to sustain in the business. 
    The E-Commerce company has the following rule for computing their own revenue: </font> 
    
####     The company charges  
####  (i) 25% on the orders having final price (discounted price) greater than 600

#### (ii) 15% on the orders having final price (discounted price) greater than 350 but less than or equal to 600

#### (iii) 10% on the orders having final price (discounted price) greater than 100 but less than or equal to 350

#### (iv) Otherwise, 5% on the final price (discounted price)    </font> 

## Q 5. Find the Total Revenue generated by the E-Commerce company over all orders placed. (6 marks)

#### Hint: Calculate revenue of E-commerce company using the conditions mentioned above and then do the total (sum) of all to get total Revenue. 

```python
#Calculating revenue as per conditions
def calculate_revenue(fp):
    if fp > 600:          #cond1
        return fp * 0.25
    elif 350 < fp <= 600: #cond2
        return fp * 0.15
    elif 100 < fp <= 350: #cond3
        return fp * 0.10
    else:                 #cond4
        return fp * 0.05
    
#applying the functions tp discounted_price column
ECom_Data['Revenue'] = ECom_Data['discounted_price'].apply(calculate_revenue)

print("Total Revenue generated by the E-Commerce company:", ECom_Data['Revenue'].sum())
```

## Q6.Calculate the total BrandRevenue and list the top 10 Brand having maximum revenue in descending order (6 marks)
#### Hint: Total BrandRevenue is BrandRevenue which is generated after all the deductions. Also, BrandRevenue is different from the E-Commerce company's revenue.
#### Brand Revenue is the discounted price minus the Revenue (generated using the conditions given above Q5)


```python
# Calculating Brand_revenue 
ECom_Data['Brand_Revenue'] = ECom_Data['discounted_price'] - ECom_Data['Revenue']

#total Brand_revenue per each brand
brand_revenue = ECom_Data.groupby('Brand')['Brand_Revenue'].sum()

#top 10 brand
top10brands = brand_revenue.sort_values(ascending=False).head(10)

print("Top 10 Brands by Brand Revenue:")
print(top10brands)
```

## Q 9.Compare performance regionwise

## a)	Draw a lineplot for the monthly Revenue of E-Commerce Company for each region separately.  (4 marks) 

## b)	Identify the best and the worst performing months for each region. (2 marks)

#### Note: Only those days with actual orders(Order_Date) placed are present in the dataset. Assuming there were no orders on other days. Also, show Month and Year on x-axis.

```python
# Extracting month and year from Order_Date
ECom_Data['Order_Date'] = pd.to_datetime(ECom_Data['Order_Date'])
ECom_Data['Month'] = ECom_Data['Order_Date'].dt.month
ECom_Data['Year'] = ECom_Data['Order_Date'].dt.year

# Group the data by Region, Year & Month
revenue_by_region = ECom_Data.groupby(['Region', 'Year', 'Month'])['Revenue'].sum().reset_index()

# Creating lineplot
plt.figure(figsize=(12, 6))
for region in revenue_by_region['Region'].unique():
    region_data = revenue_by_region[revenue_by_region['Region'] == region]
    sns.lineplot(data=region_data, x='Month', y='Revenue', hue='Year', marker='s', label=region)

plt.title('Monthly Revenue of E-Commerce Company by Region')
plt.xlabel('Month')
plt.ylabel('Revenue')
plt.xticks(range(1, 13), ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.legend(title='Year')
plt.show()
```

```python
#calculating total revenue by month for each region
revenue_by_month = ECom_Data.groupby(['Region', 'Month'])['Revenue'].sum().reset_index()

best_worst_months = {} #Empty datafram to store the worst and best month

for region in revenue_by_month['Region'].unique():
    region_data = revenue_by_month[revenue_by_month['Region'] == region]
    best_month = region_data.loc[region_data['Revenue'].idxmax(), 'Month'] #getting index of the row and values in Month column where the maximum revenue occurs by using the inbuild function idxmax()
    worst_month = region_data.loc[region_data['Revenue'].idxmin(), 'Month'] #getting index of the row and values in Month column where the maximum revenue occurs by using the inbuild function idxmin()
    best_worst_months[region] = (best_month, worst_month)

#Creating datafram so that month number and month name are connected
month_names = {
    1: 'January', 
    2: 'February', 
    3: 'March', 
    4: 'April', 
    5: 'May', 
    6: 'June',
    7: 'July', 
    8: 'August', 
    9: 'September', 
    10: 'October', 
    11: 'November', 
    12: 'December'
}

#printing the month name using the dictionary created
for region, (best_month, worst_month) in best_worst_months.items():
    print("For region",region,":")
    print("Best performing month:", month_names[best_month])  #print the month name using the dic and best_month number
    print("Worst performing month:", month_names[worst_month])  #print the month name using the dic and wrost_month number
    print()
```
